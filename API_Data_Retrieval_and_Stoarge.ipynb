```python
import pandas as pd
import numpy as np
import requests
```

```python
url = "https://openlibrary.org/search.json?q=python"

response = requests.get(url)

print(response.status_code)
```

```python
data = response.json()

print(data.keys())
```

```python
books = data["docs"]

print(books[0])
```

```python
import sqlite3
```

```python
conn = sqlite3.connect("books.db")

cursor = conn.cursor()
```

```python
cursor.execute("""
CREATE TABLE IF NOT EXISTS books (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    title TEXT,
    author TEXT,
    publication_year INTEGER
)
""")

conn.commit()
```

```python
for book in books[:20]:

    title = book.get("title", "Unknown Title")

    authors = ", ".join(
        book.get("author_name", ["Unknown Author"])
    )

    publication_year = book.get(
        "first_publish_year",
        None
    )

    cursor.execute(
        """
        INSERT INTO books
        (title, author, publication_year)
        VALUES (?, ?, ?)
        """,
        (title, authors, publication_year)
    )

conn.commit()
```

```python
cursor.execute("SELECT * FROM books LIMIT 10")

rows = cursor.fetchall()

for row in rows:
    print(row)
```
